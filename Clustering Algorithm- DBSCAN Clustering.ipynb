```python
# Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading the dataset

dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
# Getting the required input data

x = dataset.iloc[:,[3,4]].values
x
```

```python
# Standardising the input data

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
Scaled_x = sc.fit_transform(x)
```

```python
# Creating the DBSCAN model

#from sklearn.model_selection import GridSearchCV
#param_grid = {'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'], 'eps': [3,2], 'min_samples': [2]}
from sklearn.cluster import DBSCAN
#cluster_DBSCAN = GridSearchCV(DBSCAN(), param_grid, refit = True, verbose = 3, n_jobs = None)
cluster_DBSCAN = DBSCAN(eps = 0.5, min_samples = 5)
y_pred = cluster_DBSCAN.fit_predict(Scaled_x)
```

```python
# Find the Number of Clusters

labels = np.unique(y_pred)
n_clusters = len(labels)-1
print("Number of Cluster: ", n_clusters)
```

```python
# Evaluating the  cluster performance using Silhouette Score

from sklearn.metrics import silhouette_score
Silhouette_Score = silhouette_score(Scaled_x, y_pred)
print("The Silhouette Score is: ", Silhouette_Score)
```

```python
# Creating the table with cluster

dataset["Cluster_Group"] = y_pred
dataset
```

```python
# Visualixzing the Cluster Using Scatter Plot

colors = np.array(['red','blue'])
plt.scatter(Scaled_x[:,0], Scaled_x[:,1], c = colors[y_pred], s = 40, marker = 'o')
plt.title('DBSCAN Clustering')
plt.xlabel('Annaula Income')
plt.ylabel('Spending Score')
plt.show()
```
