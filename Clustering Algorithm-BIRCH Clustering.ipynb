```python
# Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the dataset

dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
# Getting the input data

x = dataset.iloc[:, [3,4]].values
x
```

```python
# Visualizing the datapoints

plt.figure(figsize = (6,4))
plt.scatter(x[:,0], x[:,1], s = 15, marker = 'x')
plt.title('Visualizing Datapoints')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# Creating the BIRCH Clustering Model

from sklearn.cluster import Birch
cluster_brich = Birch(n_clusters = 5)
y_pred = cluster_brich.fit_predict(x)
y_pred
```

```python
dir(cluster_brich)
```

```python
# Getting the Threshold

threshold = cluster_brich.threshold
print("The Threshold is: ", threshold)
```

```python
# Visualizing the clusters obtainted using BIRCH

plt.figure(figsize = (6,4))
plt.scatter(x[:,0], x[:,1], c = y_pred, cmap = 'viridis', s = 50, marker = '.')
plt.title('Mall Customers Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
