```python
# Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the dataset
 
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
# Getting the requred input data

x = dataset.iloc[:,[3,4]].values
x
```

```python
# Choosing the required number of Clusters

n_cluster = 4
```

```python
# Creating the Sperical Clustering Model

from sklearn.cluster import SpectralClustering

spectral_cluster = SpectralClustering(n_clusters = n_cluster, eigen_solver = 'arpack', random_state = 42,
                                      affinity = 'nearest_neighbors', eigen_tol = 0.0)
y_pred = spectral_cluster.fit_predict(x)
y_pred
```

```python
# Viewing the clusters using scatter plot

plt.scatter(x[:,0], x[:,1], marker = '.', s = 15, color = 'y')
plt.show()
```

```python
dir(spectral_cluster)
```

```python
affinity_matrix = spectral_cluster.affinity_matrix_
print("Affinity Matrix is: \n", affinity_matrix)
```

```python
# Evaluating the Clustering performance using silhouette score

from sklearn.metrics import silhouette_score
Silhouette_score = silhouette_score(x, y_pred, metric="sqeuclidean")
print("The Silhouette_score is:", Silhouette_score)
```

```python
# Creating the table with cluster value

dataset['cluster_group'] = y_pred
dataset
```

```python
# Visualizing the cluster

fig = plt.figure()

colors = np.array(['red', 'green', 'yellow', 'blue'])
plt.scatter(x[:,0], x[:,1], marker = '.', s = 35, c = colors[y_pred])
plt.title('Spectral Clustering of Mall Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
