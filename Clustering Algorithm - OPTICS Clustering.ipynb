```python
# importing the required libraries

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Reading the file

dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
# Getting the input data coloumns

x = dataset.iloc[:,[3,4]].values
x
```

```python
# Creating the OPTICS clustering model

from sklearn.cluster import OPTICS
cluster_optics = OPTICS(min_samples = 5, cluster_method = 'xi', algorithm='auto')
y_pred = cluster_optics.fit_predict(x)
y_pred
```

```python
dir(cluster_optics)
```

```python
# Identifying the number of clusters

labels = cluster_optics.labels_
n_clusters = len(np.unique(labels))-1
print("No. of clusters are: ",n_clusters)
```

```python
# Evaluating the cluster using the silhouette score

from sklearn.metrics import silhouette_score
s_score = silhouette_score(x,y_pred)
s_score
```

```python
# Creating the table with the cluster output

dataset['Cluster_group'] = y_pred
dataset
```

```python
# Viewing the clusters

colors = plt.cm.rainbow(np.linspace(0, 1, n_clusters))
plt.scatter(x[:, 0], x[:, 1], c=colors[y_pred], s=40, marker='o')
plt.title('OPTICS Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
colors = ['red','green','blue','yellow','purple','maroon','orange','teal','violet','pink','olive','tan','cyan','#CD7F32']
plt.figure(figsize=(6,6))
plt.scatter(x[:,0], x[:,1], c= y_pred, s = 40, cmap = matplotlib.colors.ListedColormap(colors) )
plt.title('OPTICS Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
